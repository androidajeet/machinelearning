# Machine Learning With Python: Linear Regression Multiple Variables

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

### Data Preprocessing: Fill NA values with median value of a column

```python
df.describe()
```

```python
df["experience"]
```

```python
df.columns = ['experience', 'written', 'interview','salary']
```

```python
df
```

```python
new_df= df.fillna({"experience":"zero",
                   "written":df.written.median()})
```

```python
new_df
```

```python
def wordtonum(x):
    if(x=="zero"):
        return 0
    elif(x=="one"):
        return 1
    elif(x=="two"):
        return 2
    elif(x=="three"):
        return 3
    elif(x=="four"):
        return 4
    elif(x=="five"):
        return 5
    elif(x=="six"):
        return 6
    elif(x=="seven"):
        return 7
    elif(x=="eight"):
        return 8
    elif(x=="nine"):
        return 9
    elif(x=="ten"):
        return 10
    elif(x=="eleven"):
        return 11
    

new_df["experience"]=new_df["experience"].apply(wordtonum)
non_string_df= new_df
non_string_df
```

```python
non_string_df.describe()
```

```python
non_string_df = non_string_df.replace({"experience":0
},np.NaN)
non_string_df
```

```python
non_string_df= non_string_df.fillna({"experience":non_string_df.written.median()
                   })
non_string_df
```

```python
reg = linear_model.LinearRegression()
reg.fit(non_string_df.drop('salary',axis='columns'),non_string_df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2, 9, 6]])
```
