```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('train_ctrUa4K.csv')
df.head(10)
```

```python
df["Gender"] =  df["Gender"].map({ "Male":0, "Female":1})
df["Married"] =  df["Married"].map({"Yes":1,"No":0})
df = df.replace({'Dependents': '[!@#$%^&*+]'},'', regex=True)
df["Education"] =  df["Education"].map({"Graduate":1,"Not Graduate":0})
df["Self_Employed"] =  df["Self_Employed"].map({"Yes":1,"No":0})
df["Property_Area"] =  df["Property_Area"].map({"Urban":1,"Semiurban":2,"Rural":3})
df["Loan_Status"] =  df["Loan_Status"].map({"Y":1,"N":0})

df[['Gender', 'Married',"Dependents","Education","Self_Employed","Credit_History","Property_Area","Loan_Status"]] = df[['Gender', 'Married',"Dependents","Education","Self_Employed","Credit_History","Property_Area","Loan_Status"]].fillna(method="ffill")
#df

df.head(10)
```

```python
df.describe()
```

```python
df= df.fillna({"ApplicantIncome":df.ApplicantIncome.median()})
df= df.fillna({"CoapplicantIncome":df.CoapplicantIncome.median()})
df= df.fillna({"LoanAmount":df.LoanAmount.median()})
df= df.fillna({"Loan_Amount_Term":df.Loan_Amount_Term.median()})
df
```
