```python
import pandas as pd
```

```python
df = pd.read_csv("carprices.csv")
df.head()
```

```python
#change the name of colums
df.columns = ['Car_Model', 'Mileage','Sell_Price','Age']
df.head()
```

## Using pandas to create dummy variables

```python
dummies = pd.get_dummies(df.Car_Model)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car_Model'], axis='columns')
final
```

### Dummy Variable Trap
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

<b> NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
final = final.drop(['Mercedez Benz C class'], axis='columns')
final
```

```python
X = final.drop('Sell_Price', axis='columns')
X.values
```

```python
y = final.Sell_Price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

<b>Price of mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,0,0]])
```

<b>Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```

## Using sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.head()
```

```python
dfle.Car_Model = le.fit_transform(dfle.Car_Model)
dfle.head()
```

```python
X = dfle[['Car_Model','Mileage','Age']].values
X
```

```python
y = dfle.Sell_Price.values
y
```

<b>Now use one hot encoder to create dummy variables for each of the car model

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Car_Model', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,0,45000,4]])
```

```python
model.score(X,y)
```

```python
model.predict([[1,0,86000,7]])
```
